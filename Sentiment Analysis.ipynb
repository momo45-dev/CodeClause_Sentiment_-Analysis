```python
import pandas as pd
import numpy as np 
from textblob import TextBlob
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.cluster import KMeans
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import mean_squared_error
from sklearn.svm import LinearSVC
from sklearn.model_selection import cross_val_predict
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
import scipy
```

```python
df=pd.read_csv(r'D:\Desktop\C++\New folder\Datafiniti_Hotel_Reviews_Jun19.csv')
df
```

```python
df.describe()
```

```python
 

def sentiment(label):
    if label <0:
        return "Negative"
    elif label <=0.2:
        return "Neutral"
    elif label>0.2:
        return "Positive"

```

```python
def polarity(text):
    return TextBlob(text).sentiment.polarity
```

```python
df['polarity'] = df['reviews.text'].apply(polarity)
df.head(10)
```

```python
df['sentiment'] = df['polarity'].apply(sentiment)
df
```

```python
y=df.iloc[:,2]
X=df.iloc[:,0]

```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 42) 
```

```python
X_test[0]
```

```python
## Bags_of_words
```

```python

vect=CountVectorizer()
vect_x_train = vect.fit_transform(X_train)
vect_x_test = vect.transform(X_test)
X_test=vect.transform(X_test)

```

```python
vect_x_train.shape
```

# Classification


# Linear SVC

```python
SVCmodel = LinearSVC()
SVCmodel.fit(vect_x_train, y_train)
LinearSVC()
```

```python
svc_pred = SVCmodel.predict(vect_x_test)
print(classification_report(y_test, svc_pred))
```

```python
model = KNeighborsClassifier(n_neighbors=3)

# Train the model using the training sets
model.fit(vect_x_train,y_train)

#Predict Output
predicted= model.predict(vect_x_test) # 0:Overcast, 2:Mild

print(classification_report(y_test, predicted))
```

```python
from sklearn.tree import DecisionTreeClassifier
clf_dec = DecisionTreeClassifier()
clf_dec.fit(vect_x_train,y_train)

cld_pred=clf_dec.predict(vect_x_test)

print(classification_report(y_test, cld_pred))

```

```python
logreg = LogisticRegression(multi_class='multinomial', solver='lbfgs')
logreg.fit(vect_x_train, y_train)
logreg_pred = logreg.predict(vect_x_test)
print(classification_report(y_test, logreg_pred))
```

```python
parameters = {
    'penalty' : ['l1','l2'], 
    'C'       : np.logspace(-3,3,7),
   
}
```

```python
clf_grid = GridSearchCV(logreg,                    
                   param_grid = parameters,   
                   scoring='accuracy',        
                   cv=10)        
```

```python
clf_grid.fit(vect_x_train, y_train)
```

```python
print("Best parameter:", clf_grid.best_params_)
```

```python
logreg2 = LogisticRegression(C=1,penalty="l2")
logreg2.fit(vect_x_train, y_train)
logreg_pred = logreg2.predict(vect_x_test)
print(classification_report(y_test, logreg_pred))
```
